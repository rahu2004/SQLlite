```python
import sqlite3
```

create database connection

```python
conn = sqlite3.connect('student_management.db')

cursor = conn.cursor()

print("Database Created Successfully")
```

Create SQL Queries

```python
sql_script = """


-- CREATE STUDENTS TABLE


CREATE TABLE Students (
    student_id INTEGER PRIMARY KEY,
    student_name VARCHAR(100),
    department VARCHAR(50),
    age INTEGER,
    city VARCHAR(50)
);


-- CREATE COURSES TABLE


CREATE TABLE Courses (
    course_id INTEGER PRIMARY KEY,
    course_name VARCHAR(100),
    faculty_name VARCHAR(100)
);

-- CREATE ENROLLMENTS TABLE


CREATE TABLE Enrollments (
    enrollment_id INTEGER PRIMARY KEY,
    student_id INTEGER,
    course_id INTEGER,
    marks INTEGER,

    FOREIGN KEY(student_id)
    REFERENCES Students(student_id),

    FOREIGN KEY(course_id)
    REFERENCES Courses(course_id)
);


-- INSERT STUDENTS


INSERT INTO Students VALUES
(1, 'Rahul', 'AI', 21, 'Chennai');

INSERT INTO Students VALUES
(2, 'Kiran', 'Data Science', 22, 'Hyderabad');

INSERT INTO Students VALUES
(3, 'Anjali', 'Cyber Security', 20, 'Bangalore');

INSERT INTO Students VALUES
(4, 'Rohit', 'AI', 23, 'Mumbai');


-- INSERT COURSES


INSERT INTO Courses VALUES
(101, 'Machine Learning', 'Dr. Kumar');

INSERT INTO Courses VALUES
(102, 'Data Analytics', 'Dr. Priya');

INSERT INTO Courses VALUES
(103, 'Cloud Computing', 'Dr. Sharma');

-- INSERT ENROLLMENTS


INSERT INTO Enrollments VALUES
(1, 1, 101, 90);

INSERT INTO Enrollments VALUES
(2, 2, 102, 85);

INSERT INTO Enrollments VALUES
(3, 3, 103, 88);

INSERT INTO Enrollments VALUES
(4, 4, 101, 95);

"""
```

Execute SQL Script

```python
cursor.executescript(sql_script)

conn.commit()

print("Tables and Data Created Successfully")
```

View Students Table

```python
query = "SELECT * FROM Students"

result = cursor.execute(query)

for row in result:
    print(row)
```

INNER JOIN Query

```python
query = """

SELECT
    Students.student_name,
    Courses.course_name,
    Enrollments.marks

FROM Enrollments

INNER JOIN Students
ON Students.student_id = Enrollments.student_id

INNER JOIN Courses
ON Courses.course_id = Enrollments.course_id

"""

result = cursor.execute(query)

for row in result:
    print(row)
```

Average Marks by Department

```python
query = """

SELECT
    department,
    AVG(marks) AS average_marks

FROM Students

INNER JOIN Enrollments
ON Students.student_id = Enrollments.student_id

GROUP BY department

"""

result = cursor.execute(query)

for row in result:
    print(row)
```

Highest Scoring Student

```python
query = """

SELECT
    Students.student_name,
    MAX(Enrollments.marks)

FROM Students

INNER JOIN Enrollments
ON Students.student_id = Enrollments.student_id

"""

result = cursor.execute(query)

for row in result:
    print(row)
```

Create View

```python
view_query = """

CREATE VIEW Student_Report AS

SELECT
    Students.student_name,
    Students.department,
    Courses.course_name,
    Enrollments.marks

FROM Students

INNER JOIN Enrollments
ON Students.student_id = Enrollments.student_id

INNER JOIN Courses
ON Courses.course_id = Enrollments.course_id

"""

cursor.execute(view_query)

conn.commit()

print("View Created Successfully")
```

display view

```python
query = "SELECT * FROM Student_Report"

result = cursor.execute(query)

for row in result:
    print(row)
```
